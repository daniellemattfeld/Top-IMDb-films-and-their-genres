```python
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('/Users/daniellemattfeld/Downloads/imdb_movies.csv')
```

```python
movies
```

```python
movies.info()
```

Shaping my data


```python
movies[movies.duplicated()]
```

```python
movies.drop_duplicates(inplace=True)
```

```python
movies[movies.duplicated()]
```

```python
movies.dropna(subset= ['genres'], inplace=True)
```

```python
movies.info()
```

Adding in profit column, to find highest profit film

```python
movies['profit'] = movies['revenue'] - movies['budget']
movies.head()
```

Continuing shaping/cleaning data

```python
movies_genres = movies[['popularity', 'budget', 'revenue', 'original_title','runtime', 'genres', 'release_date','vote_count','vote_average', 'profit']]
```

```python
movies_genres
```

```python
genre_split = movies_genres['genres'].str.split('|').apply(Series, 1).stack()
genre_split.index = genre_split.index.droplevel(-1)
```

```python
genre_split
```

```python
genre_split.name = 'genre_split'
```

```python
del movies_genres['genres']
```

```python
movies_genres = movies_genres.join(genre_split)
```

```python
movies_genres
```

Finding the most common genres within this data set

```python
genres_count = pd.DataFrame(movies_genres.groupby('genre_split').original_title.nunique()).sort_values('original_title', ascending= True)
```

```python
genres_count
```

```python
genres_count['original_title'].plot.pie(title = 'Movies Per Genre in %', autopct = '%1.1f%%', figsize = (10,10))
```

```python
genres_count['original_title'].plot.barh(title = 'Movies Per Genre', figsize= (10,9))
```

```python
genres_count
```

```python
genre_avg = movies_genres.groupby('genre_split').mean(numeric_only=True)
pd.options.display.float_format = '{:2f}'.format
genre_avg
```

```python
genre_avg.sort_values('budget', ascending=True, inplace=True)
```

```python
genre_avg[['budget', 'revenue']] .plot.barh(title = 'Budget and Revenue by Genre', color= ('DarkBlue', 'c'),figsize= (10,9))
```

In finding which genres have the highest average budget and revenue, I used this bar graph with both charted out. I discovered adventure, action, and fantasy, had the highest budget, while adventure, fantasy, and animation, had the highest revenue. In both, adventure ranks the highest. 

```python
movies_genres
```

```python
genre_avg.sort_values('profit', ascending=True, inplace=True)
genre_avg['profit'].plot.barh(title = 'Average Profit by Genre', color= ('DarkBlue', 'c'),figsize= (10,9))
```

In finding which genre had the highest average profit, it seems to be simliar to the revenue. The top three were adventure, fantasy, animation. 

```python
genre_avg.sort_values('popularity', ascending=True, inplace=True)
genre_avg['popularity'].plot.barh(title = 'Popularity by Genre', color= ('DarkBlue', 'c'),figsize= (10,9))
```

In finding the most popular genre, with no surprise, adventure came in first. In conclusion, with budget, revenue, profit, and popularity, films in the adventure genre rank the highest.
